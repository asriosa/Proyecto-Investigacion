```python
import pandas as pd
```

```python
manzanas = pd.read_excel('CPV 2017 - MZA agrupados.xlsx', sheet_name = 'Poblacion')
manzanas.head(10)
```

```python
manzanas2 = manzanas.iloc[7:,1:6].copy() #Solo lo que nos interesa por ahora
columnas = ['IDENTIFICADOR DE MANZANA', 'NOMBRE DEL DEPARTAMENTO', 'NOMBRE DE LA PROVINCIA', 'NOMBRE DEL DISTRITO', 'POBLACION']
manzanas2.columns = columnas
```

```python
manzanas2.head(10)
```

```python
manzanas_lima = manzanas2.copy()
manzanas_lima = manzanas_lima[(manzanas_lima['NOMBRE DEL DEPARTAMENTO'] == 'LIMA') & (manzanas_lima['NOMBRE DE LA PROVINCIA'] == 'LIMA')]
```

```python
manzanas_lima.head(10)
```

```python
manzanas_lima.shape
```

```python
manzanas_ordenado = manzanas_lima.sort_values(by='POBLACION', ascending=False)
```

```python
manzanas_ordenado
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Asumiendo que df_ordenado está correctamente ordenado y existe
top10 = manzanas_ordenado.head(10)

# Crear un gráfico de barras horizontales
plt.figure(figsize=(10, 6))  # Ajustar el tamaño del gráfico
sns.barplot(x='POBLACION', y='NOMBRE DEL DISTRITO', data=top10)

plt.title('Top 10 Manzanas')  # Añadir título al gráfico
plt.xlabel('Población')  # Nombre para el eje X
plt.ylabel('Distrito')  # Nombre para el eje Y
plt.show()
```

```python
suma_poblacion = manzanas_lima.groupby('NOMBRE DEL DISTRITO')['POBLACION'].sum().reset_index()
suma_poblacion_ordenada = suma_poblacion.sort_values(by='POBLACION', ascending=False)
```

```python
suma_poblacion_ordenada.head(10)
```

```python
# Crear un gráfico de barras horizontales
top10 = suma_poblacion_ordenada.head(10)

plt.figure(figsize=(10, 6))  # Ajustar el tamaño según necesidad
sns.barplot(x='POBLACION', y='NOMBRE DEL DISTRITO', data=top10)

plt.title('Suma de Población por Distrito')  # Añadir título al gráfico
plt.xlabel('Población')  # Nombre para el eje X
plt.ylabel('Distrito')  # Nombre para el eje Y
plt.show()
```

```python
# Importa libreria geopandas como gpd
import geopandas as gpd
```

```python
# Lee archivo shapefile de distritos del norte de Irlanda.
# Antes hay que asegurarse de crear la carpeta Shapefiles y subir todos los archivos de Shapefiles.zip
gdf_districts = gpd.read_file('MANZANA.shp') # Distritos del norte de Irlanda

# muestra el tipo de dato
type(gdf_districts)
```

```python
gdf_Lima = gdf_districts.copy()
gdf_Lima = gdf_Lima[(gdf_Lima.DEPARTAMEN == 'LIMA') & (gdf_Lima.PROVINCIA == 'LIMA')]
gdf_Lima.head(10)
```

```python
gdf_Lima.shape
```

```python
# plotea la informacion geometry del geodataframe
gdf_Lima.plot()
```

```python
if not isinstance(manzanas_lima, gpd.GeoDataFrame):
    manzanas_lima = gpd.GeoDataFrame(manzanas_lima)

if not isinstance(gdf_Lima, gpd.GeoDataFrame):
    gdf_Lima = gpd.GeoDataFrame(gdf_Lima)

# Utiliza el método merge de geopandas para mantener el tipo GeoDataFrame
df_Lima = gpd.GeoDataFrame(pd.merge(manzanas_lima, gdf_Lima, how='inner', left_on='IDENTIFICADOR DE MANZANA', right_on='IDMANZANA'))
```

```python
df_Lima.head(10)
```

```python
df_Lima.shape
```

```python
print(type(df_Lima))
```

```python
df_Lima['POBLACION'] = pd.to_numeric(df_Lima['POBLACION'], errors='coerce').fillna(0)
```

```python
# Plotea los paises del mundo coloreados segun la poblacion
fig, ax = plt.subplots(figsize = (15,5))
df_Lima.plot(ax=ax, column='POBLACION', legend=True)
plt.axis('off')   # no muestra los ejes
plt.show()
```
